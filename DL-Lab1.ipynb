```python

# Generate data
exp_data = np.random.exponential(scale=1.0, size=100000)
uniform_data = np.random.rand(100000)
normal_data = np.random.normal(loc=0.0, scale=1.0, size=100000)

# Plot histograms
plt.figure(figsize=(10, 6))
plt.hist(uniform_data, density=True, bins=100, histtype="step", color="blue", label="Uniform")
plt.hist(normal_data, density=True, bins=100, histtype="step", color="green", label="Normal")
plt.hist(exp_data, density=True, bins=100, histtype="step", color="red", label="Exponential")

# Axis and labels
plt.axis([-2.5, 8, 0, 1.1])
plt.legend(loc="upper right")
plt.title("Random Distributions (Uniform, Normal, Exponential)")
plt.xlabel("Value")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Create meshgrid
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

# 3D plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title("3D Plot of Z = X² + Y²")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
plt.show()
```

```python
#TASK 02

# Step 1: Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Step 2: Load dataset
df = pd.read_csv('Au_nanoparticle_dataset.csv')
```

```python
# Step 3: Create new_df with only required columns
new_df = df[["N_total", "N_bulk", "N_surface", "R_avg"]]
```

```python
# Step 4: Display first 20 samples
print("First 20 samples:")
print(new_df.head(20))
```

```python
# Step 5: Compute statistics: mean, std, and quartiles
print("\nStatistical Summary:")
print(new_df.describe().T[["mean", "std", "25%", "50%", "75%"]])
```

```python
# Step 6: Plot histograms in a 1x4 layout
fig, axes = plt.subplots(1, 4, figsize=(20, 4))
features = new_df.columns

for i, col in enumerate(features):
    axes[i].hist(new_df[col], bins=30, color='lightcoral', edgecolor='black')
    axes[i].set_title(f'Histogram of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
# Step 7: Pairplot - scatter and histograms
sns.pairplot(new_df)
plt.suptitle("Seaborn Pairplot of Selected Features", y=1.02)
plt.show()
```

```python
# Step 8: Advanced PairGrid with KDEs and Histograms
g = sns.PairGrid(new_df)

# Bivariate histogram on upper triangle
g.map_upper(sns.histplot)

# Histogram + KDE on diagonal
g.map_diag(sns.histplot, kde=True)

# Bivariate KDE on lower triangle
g.map_lower(sns.kdeplot, fill=True)

plt.suptitle("PairGrid with KDE and Histogram Plots", y=1.02)
plt.show()
```
